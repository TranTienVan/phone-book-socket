```python
import requests
import random
import string
from PIL import Image
import sqlite3
import pandas as pd
from string import ascii_uppercase, digits
from random import choice
```

```python
params = {
    "per_page": str(100),
    "page": str(1)
}
DataGithub = requests.get("https://api.github.com/users", params=params).json()
```

```python
def get_random_string(length):
    return ''.join(choice(ascii_uppercase) for i in range(length))

def get_random_phone(length):
    return ''.join(choice(digits) for i in range(length))
    
def random_char(char_num):
       return ''.join(random.choice(string.ascii_letters) for _ in range(char_num))

def minimize_image(url):
    image = Image.open(url)
    new_image = image.resize((200, 200))
    new_image.save(url[:-5] + "2.png")
```

```python
data = []

conn = sqlite3.connect('phone-book.db')

for user in DataGithub:
    data.append({
        "id": get_random_string(8),
        "name": user["login"],
        "phone": get_random_phone(10),
        "email": user["login"] + "@gmail.com",
        "large_photo": "/Images/{}1.png".format(user["login"]),
        "small_photo": "/Images/{}2.png".format(user["login"])
    })
    
    save_photo = "./Images/{}1.png".format(user["login"])
    user_photo = user["avatar_url"]
    
    !curl -o {save_photo} {user_photo}
    
    minimize_image("./Images/{}1.png".format(user["login"]))
    
df = pd.DataFrame(data)
df.to_sql("member", con=conn, if_exists='append', index = False)
conn.close()
```
